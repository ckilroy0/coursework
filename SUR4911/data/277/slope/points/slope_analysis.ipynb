```python
import os
import pandas as pd
import numpy as np
from dbfread import DBF
import matplotlib.pyplot as plt

def load_dbf_file(dbf_file_path):
    # Create an empty list to store rows from the DBF file
    data = []

    # Open the DBF file and read its records
    with DBF(dbf_file_path, load=True) as dbf:
        for record in dbf:
            data.append(record)

    # Create a DataFrame from the data
    df = pd.DataFrame(data)

    return df


def calculate_normalized_frequency(df, bin_size, min_value, max_value):
    # Create bins using numpy
    bins = np.arange(min_value, max_value + bin_size, bin_size)

    # Calculate the frequency distribution using pandas cut
    frequency_distribution = pd.cut(df['grid_code'], bins=bins, include_lowest=True).value_counts().sort_index()

    # Normalize the frequencies by the number of rows
    normalized_frequencies = frequency_distribution / len(df)

    # Extract the upper bound of each bin
    bin_upper_bounds = [interval.right for interval in normalized_frequencies.index]

    return bin_upper_bounds, normalized_frequencies


# Specify the directory where area folders are located
dbf_directory = 'E:/ArcticDEM/Imagery/21_38/'


# Lists to store data for plotting
all_Original_FID = []
all_bin_upper_bounds = []
all_normalized_frequencies = []
distributions = []


# Iterate over directories in the dbf_directory
for dir_name in os.listdir(dbf_directory):
    if os.path.isdir(os.path.join(dbf_directory, dir_name)):
        # Construct the dbf_file_path for the current area folder
        dbf_file_path = os.path.join(dbf_directory, dir_name, f'Slope_{dir_name}_Points.dbf')
        
        
        # Check if the DBF file exists for the current area
        if os.path.exists(dbf_file_path):
            # Load the data for the current area
            df = load_dbf_file(dbf_file_path)
            
            # Calculate normalized frequencies and get upper bounds
            bin_upper_bounds, normalized_frequencies = calculate_normalized_frequency(df, bin_size = 2, min_value = 0 , max_value = 90)
            distribution = normalized_frequencies.reset_index()['grid_code'].to_numpy()
            
            # Append the data to the lists
            all_Original_FID.append(dir_name)
            all_bin_upper_bounds.append(bin_upper_bounds)
            all_normalized_frequencies.append(normalized_frequencies)
            distributions.append(distribution)

            
# Plot the normalized frequencies for all areas on the same plot
for i, (bin_upper_bounds, normalized_frequencies) in enumerate(zip(all_bin_upper_bounds, all_normalized_frequencies)):
    label = f'Area {all_Original_FID[i]}'  # Add a label for each area
    plt.plot(bin_upper_bounds, normalized_frequencies, label=label)

plt.xlabel('Slope (Degrees)')
plt.ylabel('Normalized Frequencies')
plt.title('Normalized Frequency Distributions, Slopes')
plt.legend()  # Add a legend to distinguish areas
plt.legend(bbox_to_anchor=(1.05, 1),loc='upper left', borderaxespad=0.)
plt.show()
```

```python
# Calculate distribution differences
known_landslide_index = all_Original_FID.index('277')
diff = []
for i in range(len(distributions)):
    Sum_Abs_Diff = sum(abs(distributions[i] - distributions[known_landslide_index]))
    diff.append(Sum_Abs_Diff)

# Convert to dataframe and attach original FID
diff_df = pd.DataFrame(diff)
diff_df = diff_df.rename(columns={0: 'diff'})
diff_df['Original_FID'] = all_Original_FID

```

```python
# Plot top 3 most different distributions
# Sort the DataFrame by the 'diff' column in descending order and get the top 3 rows
top_rows = diff_df.sort_values(by='diff', ascending=False).head(3)

# Get the indices of the top rows
top_indices = top_rows.index

# Plot the normalized frequencies for all areas on the same plot
for i, (bin_upper_bounds, normalized_frequencies) in enumerate(zip(all_bin_upper_bounds, all_normalized_frequencies)):
    if i in top_indices:
        label = f'Area {all_Original_FID[i]}'  # Add a label for each area
        plt.plot(bin_upper_bounds, normalized_frequencies, label=label)

plt.xlabel('Slope (Degrees)')
plt.ylabel('Normalized Frequencies')
plt.title('Normalized Frequency Distributions, Slopes')
plt.legend()  # Add a legend to distinguish areas
plt.legend(bbox_to_anchor=(1.05, 1),loc='upper left', borderaxespad=0.)
plt.show()
```

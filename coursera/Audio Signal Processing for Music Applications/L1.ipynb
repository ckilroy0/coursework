# Programming Lecture 1: Python and sounds

## Data input

```python
from scipy.io.wavfile import read
```

```python
# Add system path
import sys
sys.path.append('sms-tools/software/models/')
```

```python
# Read wavefile
fs, x = read('sms-tools/sounds/flute-A4.wav')
```

```python
# print sample rate
print(fs)
```

```python
# print time length of array
print(f'The wavefile is {x.size/float(fs)} seconds long')
```

## Data plotting

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x)
plt.show()
```

```python
import numpy as np
```

```python
t = np.arange(x.size)/float(fs)
```

```python
plt.plot(t,x)
plt.show()
```

```python
# sample from array
y = x[44100:45100]
```

```python
plt.plot(y)
```

## Data manipulations

```python
# Get the maximum value
np.max(y)
```

```python
# Amplitude oscillates around the central axis
# The absolute maximum is found by taking the maximum value of Abs(y)
np.max(abs(y))
```

```python
# Sum the samples of the array
np.sum(y)
```

```python
# Sum the absolute values
np.sum(abs(y))
```

## Export data

```python
from scipy.io.wavfile import write

# Save to current directory
write('test.wav', fs, y)
```

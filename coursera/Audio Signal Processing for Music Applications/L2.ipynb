## Sinusoids

```python
import matplotlib.pylab as plt
import numpy as np
```

Generating Real valued sinusiods

```python
## Define variables
A = 0.8    #Signal amplitude
f0 = 1000    # Signal frequency
phi = np.pi/2    # Signal phase
fs = 44100    # Sample frequency
```

```python
t = np.arange(-0.002, 0.002, 1.0/fs)    # Real sinusoid time vector
print(t)
```

```python
## Generate Real sinusoid
x = A*np.cos(2*np.pi*f0*t + phi)
```

```python
## Plot Real sinusoid
plt.plot(t, x)
plt.axis([-0.002, 0.002, -0.8, 0.8])
plt.xlabel('time')
plt.ylabel('amplitude')
plt.show()
```

## DFT

```python
import numpy as np
```

Calculate DFT for Complex valued sinewave

```python
## Initialize X array
X = np.array([])

## Initialize n array
n = np.arange(N)
```

```python
## Calculate input signal
N = 64
k0 = 7
x = np.exp(1j*2*np.pi*k0*n/N)
```

```python
## Calculate DFT
## Calculation follows directly from definition
for k in range(N):
    s_k = np.exp(1j*2*np.pi*k*n/N)
    xs_k = x*np.conjugate(s_k)
    sum_xs_k = sum(xs_k)
    X = np.append(X, sum_xs_k)
```

```python
X
```

```python
plt.plot(np.real(X), label = 'real component')
plt.plot(np.imag(X), label = 'imaginary component')
plt.xlabel('frequency')
plt.ylabel('amplitude')
plt.legend()
plt.show()
```

```python
## Get maximum frequency
np.where(X==np.max(X))[0][0]
```

```python
## Create polar plot for Complex sinewave
r = np.sqrt(np.power(np.imag(X), 2) + np.power(np.real(X), 2))    # Magnitude
theta = np.arctan2(np.imag(X), np.real(X))    # Angle

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
plt.show()
```

```python
def genSine(A = 1.0, f = 10.0, phi = 1.0, fs = 50.0, t = 0.1):    
    """
    Inputs:
        A (float) =  amplitude of the sinusoid
        f (float) = frequency of the sinusoid in Hz
        phi (float) = initial phase of the sinusoid in radians
        fs (float) = sampling frequency of the sinusoid in Hz
        t (float) =  duration of the sinusoid (is second)
    Output:
        The function should return a numpy array
        x (numpy array) = The generated sinusoid (use np.cos())
    """
    ## Your code here
    
    # Calculate sampling period
    T = 1.0/fs
    
    # Generate time vector
    n = np.arange(0, t, T)
    
    ## Generate sinusoid
    x = A*np.cos(2*np.pi*f*n + phi)
    
    return x
```

### A2Part2

```python
def DFT(x):
    """
    Input:
        x (numpy array) = input sequence of length N
    Output:
        The function should return a numpy array of length N
        X (numpy array) = The N point DFT of the input sequence x
    """
    ## Your code here
    
    ## Initialize X array
    X = np.array([])
    
    ## Initialize n array
    N = len(x)
    n = np.arange(N)
    
    ## Calculate DFT
    ## Calculation follows directly from definition
    for k in range(N):
        s_k = np.exp(1j*2*np.pi*k*n/N)
        xs_k = x*np.conjugate(s_k)
        sum_xs_k = sum(xs_k)
        X = np.append(X, sum_xs_k)
        
    return X
```

```python
x = np.array([1, 2, 3, 4])
DFT(x)
```

```python
def IDFT(X):
    """
    Input:
        X (numpy array) = frequency spectrum (length N)
    Output:
        The function should return a numpy array of length N 
        x (numpy array) = The N point IDFT of the frequency spectrum X
    """
    ## Your code here
    
    ## Initialize signal
    x = np.array([])
    
    ## Initialize n array
    N = len(X)
    nv = np.arange(N)
    kv = np.arange(0, N)
    
    ## Calculate IDFT
    for n in nv:
        s = np.exp(1j * 2 * np.pi * n * kv / N)
        Xs_n = X*s
        sum_Xs_n = sum(Xs_n) / N    # Sum and divide by normalization factor
        x = np.append( x, sum_Xs_n)
        
    return x
```

```python
def genMagSpec(x):
    """
    Input:
        x (numpy array) = input sequence of length N
    Output:
        The function should return a numpy array
        magX (numpy array) = The magnitude spectrum of the input sequence x
                             (length N)
    """
    ## Your code here
    """
    The magnitude of a complex spectrum X is obtained by taking its absolute value
    """
    
    ### Calculate X
    ## Initialize X array
    X = np.array([])
    
    ## Initialize n array
    N = len(x)
    n = np.arange(N)
    
    ## Calculate DFT
    ## Calculation follows directly from definition
    for k in range(N):
        s_k = np.exp(1j*2*np.pi*k*n/N)
        xs_k = x*np.conjugate(s_k)
        sum_xs_k = sum(xs_k)
        X = np.append(X, sum_xs_k)
        
    ### Take absolute value of X
    magX = np.abs(X)
        
    return magX
```

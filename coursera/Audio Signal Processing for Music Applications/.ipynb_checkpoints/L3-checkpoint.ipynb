## Fourier Properties

### Basic plotting of DFT

```python
import numpy as np
from scipy.signal import triang
from scipy.fftpack import fft
```

```python
# Create signal
signal = triang(15)
```

```python
plt.plot(signal)
```

```python
# Perform FFT
DFT_signal = fft(signal)

# Calculate magnitude and phase of DFT
magnitude_DFT = abs(DFT_signal)
phase_DFT = np.angle(DFT_signal)
```

```python
# Plot DFT, magnitude and phase of DFT
fig, (ax1, ax2, ax3) = plt.subplots(3,1)
ax1.plot(DFT_signal)
ax2.plot(magnitude_DFT)
ax3.plot(phase_DFT)

plt.show()
```

### Centering around zero

```python
# Generate signal
signal = triang(15)

# Center signal around zero
fftbuffer = np.zeros(15)    # Create empty array
fftbuffer[:8] = signal[7:]    # First part of buffer is second part of signal
fftbuffer[8:] = signal[:7]    # Second part of buffer is first part of signal

# Calculate DFT
DFT_signal = fft(fftbuffer)

# Calculate magnitude and phase of DFT
magnitude_DFT = abs(DFT_signal)
phase_DFT = np.angle(DFT_signal)
```

```python
plt.plot(fftbuffer)
```

```python
# Plot DFT, magnitude and phase of DFT
fig, (ax1, ax2, ax3) = plt.subplots(3,1)
ax1.plot(DFT_signal)
ax2.plot(magnitude_DFT)
ax3.plot(phase_DFT)

plt.show()
```

The magnitude has not changed but the phase has changed. The phase is zero as a result.

### Reading sound files

```python
import sys, os, math
#sys.path.append(os.path.join(os.path.dirname(os.path.realpath(__file__)), 'sms-tools/software/models/'))
sys.path.append('sms-tools/software/models/')
import utilFunctions as UF
```

```python
# Set middle of window
M = 501    # Window size
hM1 = int(math.floor((M+1)/2))    # Start of second half   0--hM2-hM1<--M
hM2 = int(math.floor(M/2))    # End of first half   0-->hM2-hM1--M
```

```python
# Read input signal
(fs, signal) = UF.wavread('sms-tools/sounds/soprano-E4.wav')
```

```python
# Sample signal of size M
sample = signal[5000:5000+M] * np.hamming(M)
```

```python
# Center sample around zero
N = 1024
fftbuffer = np.zeros(N)
fftbuffer[:hM1] = sample[hM2:]
fftbuffer[N-hM2:] = sample[:hM2]
```

```python
fig, (ax1, ax2) = plt.subplots(2,1)
ax1.plot(sample)
ax2.plot(fftbuffer)

plt.show()
```

Note: zero buffering

```python
DFT_sample = fft(fftbuffer)
magnitude_DFT = abs(DFT_sample)
magnitude_DFT_dB = 20*np.log10(abs(DFT_sample))
phase_DFT = np.angle(DFT_sample)
```

```python
# Plot DFT, magnitude and phase of DFT
fig, (ax1, ax2, ax3) = plt.subplots(3,1)
ax1.plot(DFT_sample)
ax2.plot(magnitude_DFT_dB)
ax3.plot(phase_DFT)

plt.show()
```

```python
# Plot DFT, magnitude and phase of DFT
# Using only first half
fig, (ax1, ax2, ax3) = plt.subplots(3,1)
ax1.plot(DFT_sample[0:hM2])
ax2.plot(magnitude_DFT_dB[0:hM2])
ax3.plot(phase_DFT[0:hM2])

plt.show()
```

```python
# Plot DFT, magnitude and phase of DFT
# Using only first half
# Using unwrapped phase spectrum
fig, (ax1, ax2, ax3) = plt.subplots(3,1)
ax1.plot(DFT_sample[0:hM2])
ax2.plot(magnitude_DFT_dB[0:hM2])
ax3.plot(np.unwrap(phase_DFT[0:hM2]))

plt.show()
```

## Testing dftModel

```python
import numpy as np
from scipy.signal import get_window
import os, sys
sys.path.append('sms-tools/software/models/')
import utilFunctions as UF
import dftModel as DFT
```

```python
(fs, signal) = UF.wavread('sms-tools/sounds/piano.wav')
```

```python
M = 511
window = get_window('hamming', M)
```

```python
time = 0.2
sample = signal[int(time*fs):int(time*fs)+M]
```

```python
N = 1024
magnitude_DFT, phase_DFT = DFT.dftAnal(sample, window, N)
```

```python
y = DFT.dftSynth(magnitude_DFT, phase_DFT, window.size)*sum(window)
```

```python
fig, ax1 = plt.subplots()
ax1.plot(sample, color='blue')
ax2 = ax1.twinx()
ax2.plot(y, color='red')
plt.title('sample (blue) and synthesis (red)')
plt.ylabel('amplitude')
plt.xlabel('time')
```

```python
plt.plot(abs(sample-y))
plt.title('difference between sample and synthesis')
```

```python
# https://matplotlib.org/stable/gallery/lines_bars_and_markers/cohere.html#sphx-glr-gallery-lines-bars-and-markers-cohere-py
dt = 0.01

fig, axs = plt.subplots(2, 1)
axs[0].plot(sample, y)
axs[0].set_xlabel('Time')
axs[0].set_ylabel('sample and synthesis')
axs[0].grid(True)

cxy, f = axs[1].cohere(sample, y, 256, 1. / dt)
axs[1].set_ylabel('Coherence')

fig.tight_layout()
plt.show()
```

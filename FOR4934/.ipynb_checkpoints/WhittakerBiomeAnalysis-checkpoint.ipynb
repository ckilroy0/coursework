# Whittaker Biome Local Variation Analysis

## Introduction

A Whittaker Biome Diagram maps average temperatures (degrees C) and annual precipitation (cm) onto biomes/ecosystems. The biomes are tropical rain forest, temperate rainforest, tropical seasonal forest and savanna, temperate deciduous forest, taiga, temperate grassland and desert, subtropical desert, and tundra. The diagram shows average temperature on the x-axis and annual precipitation on the y-axis. Biomes are plotted as areas within the graph. For example, a point with an average temperature of 10 degrees C and an annual precipitation of 150 cm is classified as a temperate deciduous forest.

As climates change, a particular location may experience a shift within one biome or between multiple biomes. How can we visualize this change?

The first thing we need is weather data. We will use the Meteostat library. Second, we will need to plot the weather data. To plot the data, we will use a bivariate visualization from the seaborn library.

## Collect weather data

```python
# Import Meteostat library and dependencies
from meteostat import Point, Hourly, Daily, Monthly, Stations
from datetime import datetime, date, time
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe
import pandas as pd
import seaborn as sns
from geopy.geocoders import Nominatim
```

```python
# Set start and end dates
n = 40
start = date(2022-n, 1, 1)
end = date(2022, 12, 31)

t1 = time(0, 0)
t2 = time(23, 59)

start = datetime.combine(start, t1)
end = datetime.combine(end, t2)
```

```python
# Set location
geolocator = Nominatim(user_agent="geoapiExercises")
location = geolocator.geocode("Kansas City MO")
print(location.address)
print((location.latitude, location.longitude))
lat = location.latitude
lon = location.longitude
point = Point(lat, lon)
```

```python
#monthlyData = Monthly(station, start, end).fetch()
monthlyData = Monthly(point, start, end).fetch()
```

```python
monthlyData
```

```python
# Plot the data
plt.plot(monthlyData.index, monthlyData['prcp'])
plt.show()
```

```python
# Get year from the time index
monthlyData['year'] = monthlyData.index.year
monthlyData['month'] = monthlyData.index.month
```

```python
# Collect annual averages
grouped = monthlyData.groupby(['year']).agg({'prcp': ['sum'], 'tavg': ['mean']}).reset_index()
grouped.columns = ['year','annual_prcp', 'average_temperature']

# Round temperatures to 2 digits
grouped['average_temperature'] = round(grouped['average_temperature'],2)

# Convert precipitation from mm to cm
grouped['annual_prcp'] = grouped['annual_prcp']*0.1

# Drop NA
grouped.dropna(inplace=True)

grouped
```

```python
firstYear = grouped['year'].min()
lastYear = grouped['year'].max()
```

## Plot weather data

```python
fig, axes = plt.subplots()
sns.kdeplot( x=grouped["average_temperature"], y=grouped["annual_prcp"],  ax=axes)
sns.scatterplot(x=grouped["average_temperature"],y=grouped["annual_prcp"], ax = axes, hue=grouped["year"])
#sns.lineplot(grouped,x=grouped["average_temperature"],y=grouped["annual_prcp"], sort=False, color='black', lw=1)
sns.move_legend(axes, "upper left", bbox_to_anchor=(1, 1))
axes.invert_xaxis()

plt.title(f"Kernel Density Estimate and Scatterplot \n {location.address} \n {location.latitude, location.longitude}")

plt.xlabel('Average Temperature (C)')
plt.ylabel('Annual Precipitation (cm)')
```

The weather data has been plotted and can now be compared to the Whittaker Biome Diagram.

## Analysis

```python
import matplotlib.image as mpimg # add
#h = sns.heatmap(df_cnt, alpha=0.1, zorder=2) # update
fig = plt.figure()
axes = fig.add_subplot(111)
#fig, axes = plt.subplots()
fig.add_subplot(sns.kdeplot( x=grouped["average_temperature"], y=grouped["annual_prcp"], ax=axes, color="black"))
my_image = mpimg.imread('whittakerbiomediagram-trimmed.png') # update
# update
#axes.set_xlim(0,733)
axes.set_xlim(34.4,-19.9)
axes.set_ylim(0,450)
fig.add_subplot(sns.scatterplot(x=grouped["average_temperature"],y=grouped["annual_prcp"], ax = axes, hue=grouped["year"]))

# show the background image
x0,x1 = axes.get_xlim()
y0,y1 = axes.get_ylim()
axes.imshow(my_image, extent=[x0, x1, y0, y1], aspect='auto')

fig.set_size_inches(9, 7)

plt.title(f"Whittaker Biome Analysis \n{location.address} \n{location.latitude, location.longitude} \nAggregated Sample Size: {len(grouped)}")
plt.xlabel('Average Temperature (C)')
plt.ylabel('Annual Precipitation (cm)')
plt.legend(loc='center right', title='Year')

plt.show() # add
```

## Adding a trace to the plot

I want to see a trace of temperature versus cumulative monthly precipitation. Instead of a single point showing average temperature, the trace should show how temperature varies annually. Precipitation should converge to the annual precipitation.

```python
monthlyData
```

```python
monthlyData['Cum_prcp'] = monthlyData[['year','prcp']].groupby('year').cumsum()
monthlyData['Cum_prcp'] = monthlyData['Cum_prcp']*0.1

monthlyData2 = monthlyData
monthlyData2 = monthlyData2.sort_values(by='Cum_prcp',ascending=True)
monthlyData2['Cum_prcp'] = round(monthlyData2['Cum_prcp'],0)
monthlyData2 = monthlyData2.reset_index()
```

```python
from scipy.ndimage import gaussian_filter1d
#tavg_smoothed = pd.DataFrame(gaussian_filter1d(monthlyData2['tavg'].interpolate(), sigma=2))
tavg_smoothed['tavg_smoothed_2s'] = pd.DataFrame(gaussian_filter1d(monthlyData2['tavg'].interpolate(), sigma=2))
tavg_smoothed['tavg_smoothed_10s'] = pd.DataFrame(gaussian_filter1d(monthlyData2['tavg'].interpolate(), sigma=10))
monthlyData2 = monthlyData2.join(tavg_smoothed['tavg_smoothed_2s'])
monthlyData2 = monthlyData2.join(tavg_smoothed['tavg_smoothed_10s'])
```

```python
# Collect monthly averages
grouped2 = monthlyData2.groupby(['month']).agg({'Cum_prcp': ['mean'], 'tavg': ['mean']}).reset_index()
grouped2.columns = ['month','Cum_prcp_mean', 'tavg_mean']

# Round temperatures to 2 digits
grouped2['tavg_mean'] = round(grouped2['tavg_mean'],2)

# Convert precipitation from mm to cm
grouped2['Cum_prcp_mean'] = grouped2['Cum_prcp_mean']

# Drop NA
#grouped.dropna(inplace=True)

grouped2
```

```python
fig, axes = plt.subplots()
sns.kdeplot( x=grouped["average_temperature"], y=grouped["annual_prcp"],  ax=axes)
sns.scatterplot(x=grouped["average_temperature"],y=grouped["annual_prcp"], ax = axes, hue=grouped["year"])
sns.lineplot(data=monthlyData2, x="tavg", y="Cum_prcp", orient="y", errorbar=("se", 2), ax = axes,  hue='month')
sns.lineplot(data=monthlyData2, x="tavg_smoothed_10s", y="Cum_prcp", orient="y", errorbar=("se", 2), ax = axes, linestyle='dashed')
#sns.lineplot(grouped,x=grouped["average_temperature"],y=grouped["annual_prcp"], sort=False, color='black', lw=1)
sns.move_legend(axes, "upper left", bbox_to_anchor=(1, 1))
axes.invert_xaxis()

plt.title(f"Kernel Density Estimate and Scatterplot \n {location.address} \n {location.latitude, location.longitude}")

plt.xlabel('Average Temperature (C)')
plt.ylabel('Annual Precipitation (cm)')

```

```python
fig, axes = plt.subplots()
sns.kdeplot( x=grouped["average_temperature"], y=grouped["annual_prcp"],  ax=axes, label='Annual Kernel Density Estimate',zorder=1)
sns.scatterplot(x=grouped["average_temperature"],y=grouped["annual_prcp"], ax = axes, hue=grouped["year"], zorder=2)
#sns.lineplot(data=monthlyData2, x="tavg", y="Cum_prcp", orient="y", errorbar=("se", 2), ax = axes)
sns.lineplot(data=grouped2, x="tavg_mean", y="Cum_prcp_mean", orient="y", ax = axes, label='Monthly Derived Trace', lw=1.5, path_effects=[pe.Stroke(linewidth=1.7, foreground='black'), pe.Normal()])
sns.lineplot(data=monthlyData2, x="tavg_smoothed_10s", y="Cum_prcp", orient="y", errorbar=("se", 2), ax = axes, label='Accumulate-Derived Trace', lw=1.5, path_effects=[pe.Stroke(linewidth=1.7, foreground='black'), pe.Normal()])
#sns.lineplot(grouped,x=grouped["average_temperature"],y=grouped["annual_prcp"], sort=False, color='black', lw=1)

plt.legend(loc='lower left', bbox_to_anchor=(1, 0.2))
#plt.legend(['Monthly Derived Estimate','Accumulate-Derived Estimate','Annual Kernel Density Estimate',f'Annual Estimate, {firstYear}',f'Annual Estimate, {lastYear}'])

axes.invert_xaxis()

plt.title(f"Kernel Density Estimate and Scatterplot \n {location.address} \n {location.latitude, location.longitude}")

plt.xlabel('Average Temperature (C)')
plt.ylabel('Annual Precipitation (cm)')

```
